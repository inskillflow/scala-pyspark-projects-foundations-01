```python
import pyspark
from pyspark import SparkContext
from pyspark.sql import SparkSession
import matplotlib.pyplot as plt
```

```python

spark = SparkSession.builder.master("local").appName("Apple store market Analysis").getOrCreate()
```

```python
StocksAAMEDF = spark.read.option("header",True) \
     .csv("AAME.csv")
```

```python
StocksAAMEDF.show()
```

```python
StocksAAMEDF.createOrReplaceTempView("AAME");
```

```python
spark.sql("select * from AAME").show()
```

```python
spark.sql("select Date,Open, High,Low,Close,'Adj Close',Volume from AAME").show()
```

```python
spark.sql("select Date as dt, (Close-Open) as DifPrixOuverturePrixFermeture from AAME ").show()
```

```python
spark.sql("select Date, Open,Close from AAME").show()
```
