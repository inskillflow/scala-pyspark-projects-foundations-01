# **Question 1 - Liste des dates de transaction, ouverture et fermeture de la valeur de l’action**

# **Question 2 - Liste des dates de transaction avec la différence entre fermeture et ouverture**

# **Question 3 - Max et Min des volumes**

```python
spark.sql("select max(Volume) as max_volume, min(Volume) as min_volume from AAPL").show()
```

# **Question 4 - Moyenne des valeurs d’ouverture par années**

---



```python
spark.sql("select year(Date) as year, avg(Open) as avg_open from AAPL group by year(Date) order by year").show()
```

# **Question 5 - Somme des volumes par mois**

```python
spark.sql("select year(Date) as year, month(Date) as month, sum(Volume) as total_volume from AAPL group by year(Date), month(Date) order by year, month").show()
```

```python
spark.sql("""
    SELECT date_format(Date, 'yyyy-MM') AS annee_mois,
           SUM(Volume) AS sumvolum
    FROM AAPL
    GROUP BY date_format(Date, 'yyyy-MM')
    ORDER BY annee_mois
""").show()
```
